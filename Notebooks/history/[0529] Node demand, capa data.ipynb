```python
import pandas as pd
```

```python
demand = pd.read_csv('./../../Data/통합 문서1.CSV')
data = pd.read_csv("./../../Data/data_with_depot.csv")
```

```python
temp = pd.merge(left=data, right=demand[['latitude', 'longitude', '기기 한대로의 capa 변환값 (L)', 'demand (L)']], on=['latitude', 'longitude'], how='left')
temp.drop(columns =['Capacity'], inplace=True)
temp = temp.rename(columns={'기기 한대로의 capa 변환값 (L)' : 'capacity', 'demand (L)' : 'demand_mu'})
temp = temp.fillna(0)
```

```python
temp.to_csv("./../../Data/data_with_depot.csv", index=False)
```
