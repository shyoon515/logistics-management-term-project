```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
df = pd.read_csv('./../Data/data.csv')
```

```python
df
```

```python
df1 = df[df.machine == 1]
df2 = df[df.machine == 2]

ratio = (df.latitude.max() - df.latitude.min()) / (df.longitude.max() - df.longitude.min())
scale = 6

plt.figure(figsize=(scale, scale*ratio))
plt.scatter(df1['longitude'], df1['latitude'], s= 1.2, c='blue', label = '1 Machine')
plt.scatter(df2['longitude'], df2['latitude'], s= 1.2, c='red', label = '2 Machines')
#plt.plot(df['latitude'], df['longitude'], 'b-')
plt.xlabel('Londitude')
plt.ylabel('Latitude')
plt.title('전남 지역 네프론 기기 대수별 위치')
plt.legend()
plt.show()
```
