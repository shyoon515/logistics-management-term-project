```python
import numpy as np
import pandas as pd
```

```python
# 저장된 배열 불러오기
cost_matrix = np.load('./../Data/cost_matrix.npy')
node_demands = pd.read_csv('./../Data/node_demand.csv')
```

```python
# calculate savings for each link
savings = dict()
for i in range(len(cost_matrix)):
    for j in range(len(cost_matrix)):
        # garbage값은 skip
        if cost_matrix[i][j] == -1:
            continue
        if i != j:
            a = min(i, j)
            b = max(i, j)
            key = '(' + str(a) + ',' + str(b) + ')'
            # cost_matrix와 node_demands 데이터 생성 시 depot index가 다르면 error 발생
            savings[key] = node_demands.loc[node_demands['node'] == i, 'depot_node_cost'].values[0] + node_demands.loc[node_demands['node'] == j, 'depot_node_cost'].values[0] - int(cost_matrix[i][j])

sv = pd.DataFrame.from_dict(savings, orient = 'index')
sv.rename(columns = {0:'saving'}, inplace = True)
sv.sort_values(by = ['saving'], ascending = False, inplace = True)
sv.head()
```

```python
def get_node(link):
    link = link[1:]
    link = link[:-1]
    nodes = link.split(',')
    return [int(nodes[0]), int(nodes[1])]

def interior(node, route):
    """
    [a, b, c, d] 와 같은 route가 들어오면 node가 interior point인지 판단
    """
    try:
        i = route.index(node)
        # adjacent to depot, not interior
        if i == 0 or i == (len(route) - 1):
            label = False
        else:
            label = True
    except:
        label = False
    
    return label

def merge(route0, route1, link):
    """
    link를 기준으로 두 route를 합침
    """
    if route0.index(link[0]) != (len(route0) - 1):
        route0.reverse()
    
    if route1.index(link[1]) != 0:
        route1.reverse()
        
    return route0 + route1

# sum up to obtain the total passengers belonging to a route
def sum_cap(route):
    sum_cap = 0
    for node in route:
        sum_cap += node_demands.loc[node_demands['node'] == node, 'demand'].values[0]
    return sum_cap
```
