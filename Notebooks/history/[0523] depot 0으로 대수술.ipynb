### json 파일 이름 모두 변경

```python
all_jsons = os.listdir('./../Data/routes')
file_names = [f for f in all_jsons]

for f in file_names:
    # 정규표현식 패턴 활용
    pattern = r'route_(\d+)_(\d+)\.json'
    match = re.match(pattern, f)
    start_idx = int(match.group(1))
    end_idx = int(match.group(2))

    if start_idx > end_idx:
        os.remove('./../Data/routes/'+f)
```

```python
# 폴더 A와 폴더 B의 경로 설정
folder_A = './../Data/routes/'
folder_B = './../Data/routes_with_depot/'

# 폴더 B가 존재하지 않으면 생성
if not os.path.exists(folder_B):
    os.makedirs(folder_B)

# 폴더 A의 모든 파일을 순회
for filename in tqdm(os.listdir(folder_A)):
    # 파일의 전체 경로
    file_A = os.path.join(folder_A, filename)

    # 정규표현식 패턴 활용
    pattern = r'route_(\d+)_(\d+)\.json'
    match = re.match(pattern, filename)
    start_idx = int(match.group(1))
    end_idx = int(match.group(2))
    
    # 파일인 경우에만 처리 (디렉토리 제외)
    if os.path.isfile(file_A):
        # 새로운 파일 이름 생성 (여기서는 접두사를 추가하는 예시)
        new_filename = f"route_{start_idx+1}_{end_idx+1}.json"

        # 새로운 파일의 전체 경로
        file_B = os.path.join(folder_B, new_filename)

        # 파일 복사
        shutil.copy2(file_A, file_B)
```
